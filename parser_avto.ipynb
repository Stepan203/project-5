```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from tqdm import tqdm
import time
base_url = 'https://auto.ru/cars/'
#base_url = 'https://auto.ru/cars/all/?output_type=table'
#page_part = '&page='
page_part = 'used/?page='
cars_all = []
df_brand = ['mitsubishi']
#for j in df_brand:
    #url_gen = 
for i in range(3701, 3900):
    url_gen = base_url + page_part + str(i)
    rt = requests.get(url_gen)
    rt.encoding = 'utf-8'
    soup = BeautifulSoup(rt.text, 'html.parser')
    #ads = soup.find_all('div', class_='ListingItem-module__container oQiaM_TH7PEp_tywwiY4ayshB3r1VnqiV5nrjFNIKoA')
    ads = soup.find_all('div', class_='ListingItem')
    #print(rt)
    for ur in tqdm(ads):
        title_url = ur.find('a', class_='Link ListingItemTitle__link').get('href')
        rt = requests.get(title_url)
        rt.encoding = 'utf-8'
        soup = BeautifulSoup(rt.text, 'html.parser')
        #ads = soup.find_all('div', class_='oXi6Y7Bv0KW0zqDB5kY0b4Ptn4JBd__container')
        ads = soup.find_all('div', class_='PageCard')
        #print(len(ads))
        time.sleep(0.5)
        cars = []
        try:
            cars.append(soup.find('div', class_='CardHead__shortInfo CardHead__shortInfo_used').find_all('div', class_='CardHead__infoItem CardHead__id')[-1].text.split(' ')[-1])
        except:
            auto_id = cars.append('Nan')
        try:
            cars.append(soup.find('div', class_='CardHead__topRowLeftColumn').find_all('h1', class_='CardHead__title')[0].text.split(' ')[0])
        except:
            brand = cars.append('Nan')
        try:
            cars.append(soup.find('div', class_='CardHead__topRowLeftColumn').find_all('h1', class_='CardHead__title')[0].text)
        except:
            tl = cars.append('Nan')
        try:
            cars.append(soup.find('div', class_='CardHead__shortInfo CardHead__shortInfo_used').find_all('div', class_='CardHead__infoItem CardHead__creationDate')[-1].text)
           
        except:
            times = cars.append('Nan')
        try:    
            cars.append(soup.find('div', class_='CardHead__shortInfo CardHead__shortInfo_used').find_all('div', class_='CardHead__infoItem CardHead__views')[-1].text)
        except:
            views = cars.append('Nan')
        try:
            cars.append(soup.find('li', class_='CardInfoRow CardInfoRow_year').find('a').text)
        except:
            year = cars.append('Nan')
        try:
            cars.append(soup.find('li', class_='CardInfoRow CardInfoRow_kmAge').find_all('span', class_='CardInfoRow__cell')[-1].text)
        except:
            kmage = cars.append('Nan')
        try:
            cars.append(soup.find('li', class_='CardInfoRow CardInfoRow_bodytype').find_all('span', class_='CardInfoRow__cell')[-1].text)
        except:
            body_type = cars.append('Nan')
        try:
            cars.append(soup.find('li', class_='CardInfoRow CardInfoRow_color').find_all('span', class_='CardInfoRow__cell')[-1].text)
        except:
            color = cars.append('Nan')
        try:
            cars.append(soup.find('li', class_='CardInfoRow CardInfoRow_engine').find_all('span', class_='CardInfoRow__cell')[1].text.split('/')[0])
        except:
            engineDisplacement = cars.append('Nan')
        try:
            cars.append(soup.find('li', class_='CardInfoRow CardInfoRow_engine').find_all('span', class_='CardInfoRow__cell')[1].text.split('/')[1])
        except:
            enginePower = cars.append('Nan')
        try:
            cars.append(soup.find('li', class_='CardInfoRow CardInfoRow_engine').find_all('span', class_='CardInfoRow__cell')[1].text.split('/')[-1])
        except:
            fuelType = cars.append('Nan')
        try:
            cars.append(soup.find('li', class_='CardInfoRow CardInfoRow_transportTax').find_all('span', class_='CardInfoRow__cell')[-1].text.split('/')[0])
        except:
            tax = cars.append('Nan')
        try:
            cars.append(soup.find('li', class_='CardInfoRow CardInfoRow_transmission').find_all('span', class_='CardInfoRow__cell')[-1].text)
        except:
            transmision = cars.append('Nan')
        try:
            cars.append(soup.find('li', class_='CardInfoRow CardInfoRow_drive').find_all('span', class_='CardInfoRow__cell')[-1].text)
        except:
            drive = cars.append('Nan')
        try:
            cars.append(soup.find('li', class_='CardInfoRow CardInfoRow_wheel').find_all('span')[-1].text)
        except:
            wheel = cars.append('Nan')
        try:
            cars.append(soup.find('li', class_='CardInfoRow CardInfoRow_state').find_all('span')[-1].text)
        except:
            state = cars.append('Nan')
        try:
            cars.append(soup.find('li', class_='CardInfoRow CardInfoRow_ownersCount').find_all('span')[-1].text)
        except:
            onwers_count = cars.append('Nan')
        try:
            cars.append(soup.find('li', class_='CardInfoRow CardInfoRow_pts').find_all('span')[-1].text)
        except:
            pts = cars.append('Nan')
        try:
            cars.append(soup.find('li', class_='CardInfoRow CardInfoRow_customs').find_all('span')[-1].text)
        except:
            castoms = cars.append('Nan')
        try:
            cars.append(soup.find('div', class_='PriceUsedOffer').find_all('span', class_='OfferPriceCaption__price')[-1].text)
        except:
            price = cars.append('Nan')
        
         
        cars_all.append(cars)
        df = pd.DataFrame(cars_all, columns=['id', 'brand', 'title', 'times', 'views', 'year', 'kmage', 'body_type','color','engineDisplacement', 'enginePower', 'fuelType', 'tax', 'transmision', 'drive', 'wheel', 'state', 'onwers_count', 'pts', 'castoms', 'price'])
df.to_csv ('avto_all30.csv', encoding='utf-8')
```
